## Geany Themes Scraper

Automatic download of all official Geany IDE themes...

```python
import os

import pandas as pd
from requests_html import HTMLSession
```

```python
URL = 'https://www.geany.org/download/themes/'
session = HTMLSession()
```

```python
response = session.get(URL)
response.status_code
```

```python
themes = response.html.xpath('//div[@class="col-md-4"]')
themes
```

```python
themes_ = []

for theme in themes:
    panel_heading = theme.xpath('//div[@class="panel-heading"]', first=True)
    name = panel_heading.xpath('//strong/text()', first=True)
    url = panel_heading.xpath('//span/a/@href', first=True)

    panel_body = theme.xpath(
        '//div[contains(@class, "panel-body")]', first=True)
    image = panel_body.xpath('//a/@href', first=True)

    themes_.append({
        'Theme_Name': name,
        'Theme_URL': url,
        'Image_URL': image,
    })
```

```python
themes_
```

```python
df = pd.DataFrame(themes_)
df.sample(5)
```

```python
df['Theme_URL'].map(
    lambda x: os.system(f'wget {x} -P ./exports')
)
```

```python
df.to_csv('./exports/geany_themes.csv', index=False)
```
