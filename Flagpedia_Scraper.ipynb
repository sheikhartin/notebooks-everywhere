## Extracting Country Flag Emoji

It is about a simple project for my YouTube that I wanted to teach the basics of web scraping!

```python
import pandas as pd
from requests_html import HTMLSession
```

```python
URL = 'https://flagpedia.net/emoji'
session = HTMLSession()
```

```python
response = session.get(URL)
response.status_code
```

```python
response.content
```

```python
rows = response.html.xpath('//tbody/tr')
rows
```

```python
countries = []

for row in rows:
    emoji = row.xpath('//td[@class="td-flag"]/a/text()', first=True)
    emoji = (
        emoji
        .replace('\t', '')
        .replace('\n', '')
    )
    country = row.xpath('//td[@class="td-country"]/a/text()', first=True)
    codepoints = row.xpath(
        '//td[contains(@class, "td-codepoints")]/code/text()'
    )
    codepoints = ', '.join(codepoints)
    countries.append({
        'emoji': emoji,
        'country': country,
        'codepoints': codepoints,
    })
```

```python
countries
```

```python
df = pd.DataFrame(countries)
df
```

```python
df.to_csv('./exports/countries_flag.csv')
```
