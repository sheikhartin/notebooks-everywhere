```python
import gradio as gr
```

A simple function that can be called the second "Hello, world!" program:

```python
def greet(name: str) -> str:
    return f'Hello, {name}!'
```

```python
greet('Artin')
```

Quick demo of the `greet` function:

```python
demo = gr.Interface(fn=greet, inputs='text', outputs='text')
demo.launch(share=True)
```
